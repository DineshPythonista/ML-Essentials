```python
import numpy as np
import pandas as pd
import math as m
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
def entropy_class(df, Y):
    classes = list(df[Y].unique())
    entropy_of_class = 0.0
    for i in classes:
        b = len(df[df[Y] == i]) / len(df)
        entropy_of_class += -b * m.log2(b)
    return entropy_of_class

def entropy(df, fe_n, Y):
    classes = list(df[Y].unique())
    val = list(df[fe_n].unique())
    entropy_of_attribute = 0.0
    for i in val:
        a = df[df[fe_n] == i]
        b = len(a)
        c = b / len(df)
        each = 0.0
        for j in classes:
            d = len(a[a[Y] == j]) / b
            if d != 0:  # purity
                each += -d * m.log2(d)
        entropy_of_attribute += c * each
    return entropy_of_attribute

def gain(df, Y):
    fe_n = list(df.iloc[:, :-1])  # Get feature columns excluding the target column
    e_c = entropy_class(df, Y)
    gain = []
    for i in range(len(fe_n)):
        gain.append((e_c - entropy(df, fe_n[i], Y), i))
    gain.sort(key=lambda x: x[0], reverse=True)
    return gain[0][1]

def recursive(maxd, df):
    if maxd <= 0 or len(df['Play Golf'].unique()) == 1:
        node = {'class': df['Play Golf'].mode()[0]}
        return node

    fe_n = list(df.iloc[:, :-1])
    t = gain(df, 'Play Golf')
    split = df[fe_n[t]].unique()

    node = {'attribute': fe_n[t], 'children': {}}

    for value in split:
        new_df = df[df[fe_n[t]] == value].drop(fe_n[t], axis=1)
        subtree = recursive(maxd - 1, new_df)
        node['children'][value] = subtree

    return node

def predict_instance(instance, tree):
    if 'class' in tree:
        return tree['class']

    attribute = tree['attribute']
    value = instance[attribute]

    if value not in tree['children']:
        return None

    child_tree = tree['children'][value]
    return predict_instance(instance, child_tree)

def predict(test_data, tree):
    predictions = []
    for _,instance in test_data.iterrows():
        prediction = predict_instance(instance, tree)
        predictions.append(prediction)
    return predictions

df = pd.read_csv('golf.csv')
decision_tree = recursive(2, df)
train,test=train_test_split(df,test_size=0.2,random_state=42)
X_test=test.iloc[:,:-1]
Y_test=test.iloc[:,-1]
predictions = predict(X_test, decision_tree)
print("Predictions:", predictions)
accuracy=accuracy_score(predictions,Y_test)
print(accuracy)
import seaborn as sns
sns.heatmap(confusion_matrix(predictions,Y_test),annot=True)
```

```python
import numpy as np
import pandas as pd
import math as m
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

def entropy_class(df, Y):
    classes = list(df[Y].unique())
    entropy_of_class = 0.0
    for i in classes:
        b = len(df[df[Y] == i]) / len(df)
        entropy_of_class += -b * m.log2(b)
    return entropy_of_class


def gain(df, Y):
    fe_n = list(df.iloc[:, :-1])
    e_c = entropy_class(df, Y)
    gain = []
    for i in range(len(fe_n)):
        gain.append((e_c - median1(fe_n[i],df, Y), i))
    gain.sort(key=lambda x: x[0], reverse=True)
    return gain[0][1]

def recursive(maxd, df):
    if maxd <= 0 or len(df['Class'].unique()) == 1:
        node = {'class': df['Class'].mode()[0]}
        return node

    fe_n = list(df.iloc[:, :-1])
    t = gain(df, 'Class')
    split = median1(fe_n[t], df, Y='Class')

    node = {'attribute': fe_n[t], 'split_point': split, 'children': {}}

    new_df_left = df[df[fe_n[t]] <= split].drop(fe_n[t], axis=1)
    new_df_right = df[df[fe_n[t]] > split].drop(fe_n[t], axis=1)

    if len(new_df_left) == 0 or len(new_df_right) == 0:
        node = {'class': df['Class'].mode()[0]}
        return node

    node['children']['<='] = recursive(maxd - 1, new_df_left)
    node['children']['>'] = recursive(maxd - 1, new_df_right)

    return node

def predict_instance(instance, tree):
    if 'class' in tree:
        return tree['class']

    attribute = tree['attribute']
    value = instance[attribute]

    if value <= tree['split_point']:
        child_tree = tree['children']['<=']
    else:
        child_tree = tree['children']['>']

    if len(child_tree) == 0 :
        return tree['class']

    return predict_instance(instance, child_tree)


def predict(test_data, tree):
    predictions = []
    for _, instance in test_data.iterrows():
        prediction = predict_instance(instance, tree)
        predictions.append(prediction)
    return predictions

def median1(fe_n, data, Y):
    gain = []
    a = data[fe_n].unique()
    classes = sorted(list(data[Y].unique()))
    entropy_median = 0.0
    for i in a:
        v = data[data[fe_n] <= i]
        b = len(v)
        if b == 0 or b == len(data):
            continue
        c = b / len(data)
        each = 0.0
        for j in classes:
            d = len(v[v[Y] == j]) / b
            if d != 0:
                each += d * m.log2(d)
        entropy_median += c * each
        u = data[data[fe_n] > i]
        e = len(u)
        if e == 0 or e == len(data):
            continue
        f = e / len(data)
        eu = 0.0
        for j in classes:
            h = len(u[u[Y] == j]) / e
            if h != 0:
                eu += h * m.log2(h)
        entropy_median += f * eu
        gain.append([entropy_median, i])
    gain.sort(key=lambda x: x[0])
    return gain[0][1]


df = pd.read_csv('Date_Fruit_Datasets.csv')
decision_tree = recursive(8, df)
print(decision_tree)

```

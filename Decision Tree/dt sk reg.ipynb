```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeRegressor, export_graphviz
import seaborn as sns
import matplotlib.pyplot as plt
import graphviz
```

```python
df = pd.read_csv("Date_Fruit_Datasets.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df["Class"] = df["Class"].astype("category")
df["Class"] = df["Class"].cat.codes
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 1, test_size = 0.3)
```

```python
clf = DecisionTreeRegressor(random_state = 42)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
rmse
```

```python
graph = export_graphviz(clf, feature_names=x.columns,class_names=y.unique(),filled=True)
tree = graphviz.Source(graph,format="png")
tree
```

```python
imp = clf.feature_importances_
sorted = np.argsort(imp)

plt.barh(range(len(imp)), imp[sorted], align = "center")
plt.yticks(range(len(imp)), [x.columns[i] for i in sorted])
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.title("Decision Tree clf - Feature importance")
plt.show()
```
